```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

L_data = pd.read_csv("Data/L_data_merged.csv", low_memory = False)

L_data['date'] = pd.to_datetime(L_data['date'], infer_datetime_format=True)
L_data.head()
```

```python
L_data.loc[(L_data['date']>= "01-01-2018") & (L_data['date']<= "01-27-2019")].groupby(pd.Grouper(key='date', freq='W')).agg({
    'rides': 'sum'
}).plot(kind = 'line', color='green')

plt.xlabel('Date')
plt.ylabel('Number of Rides')
plt.title('Number of Rides vs Date (2018-2019)')
```

```python
L_data.loc[(L_data['date']>= "01-07-2019") & (L_data['date']<= "01-26-2020")].groupby(pd.Grouper(key='date', freq='W')).agg({
    'rides': 'sum'
}).plot(kind = 'line', color='b')

plt.xlabel('Date')
plt.ylabel('Number of Rides')
plt.title('Number of Rides vs Date (2019-2020)')
```

```python
L_data.loc[(L_data['date']>= "01-06-2020") & (L_data['date']<= "01-31-2021")].groupby(pd.Grouper(key='date', freq='W')).agg({
    'rides': 'sum'
}).plot(kind = 'line', color='red')

plt.xlabel('Date')
plt.ylabel('Number of Rides')
plt.title('Number of Rides vs Date (2020)')
```

```python
bus_data = pd.read_csv("Data/bus_ridership_daily.csv", low_memory = False)

bus_data['date'] = pd.to_datetime(bus_data['date'], infer_datetime_format=True)
bus_data.head()
```

```python
bus_data.loc[(bus_data['date']>= "01-07-2019") & (bus_data['date']<= "01-26-2020")].groupby(pd.Grouper(key='date', freq='W')).agg({
    'rides': 'sum'
}).plot(kind = 'line', color='green')

plt.xlabel('Date')
plt.ylabel('Number of Bus Rides')
plt.title('Number of Bus Rides vs Date (2019-2020)')
```

Analysis:

From the data it can be observed that holidays have an effect on L and bus line travel. There is a noticeable increase in ridership in mid-March every year. In Chicago St. Patrick's day is an extremely popular holiday and the date and parade correspond with the spike in rides. Furthermore, from the data is can be seen that around holidays in which individuals have time off from work or school (Thanksgiving, Christmas and New Years to name a few), the number of rides on the L and bus lines tend to go down. This is understandable as less people would be using public transport for work purposes during this time. Finally, it is observable that the trend of L and bus ridership was on a similar track in 2020 as compared to the previous two years, until the beginning of March in which there is a large drop off in travel. This can be explained by Covid-19 and the procautions that the city of Chicago and the State of Illinois put in place to reduce the spread of the pandemic.
